# **Note 1: Loaded dataset using pd.read_csv() to start analysis.**
### **Note 2: Used .head() to preview first 5 rows and understand columns**

```python
import pandas as pd

df = pd.read_csv("spotify_data clean.csv")
df.head()
```

**Note 3: Used .info() to check datatypes and non-null counts.**

```python
df.info()
```

```python
df.columns
```

**Note 4: Found missing values using .isnull().sum().**

```python
df.isnull().sum()
```

**Note 5: Filled missing artist_name with "Unknown" because it is categorical text.**

**Note 6: Filled missing artist_genres with "Not Available" to avoid losing rows. **

```python
df["artist_name"] = df["artist_name"].fillna("Unknown")
df["artist_genres"] = df["artist_genres"].fillna("Not Available")
```

```python
df.duplicated().sum()
```

```python
df_before = df.shape[0]
df = df.drop_duplicates()
df_after = df.shape[0]

print("Rows before:", df_before)
print("Rows after:", df_after)
```

```python
df["album_release_date"] = pd.to_datetime(df["album_release_date"], errors="coerce")
```

```python
df["release_year"] = df["album_release_date"].dt.year
df["duration_seconds"] = (df["track_duration_min"] * 60).round(0)
```

```python
def popularity_category(x):
    if x >= 80:
        return "Very Popular"
    elif x >= 50:
        return "Popular"
    else:
        return "Less Popular"

df["popularity_category"] = df["track_popularity"].apply(popularity_category)
```

```python
df.isnull().sum()
df.info()
df.head()
```

```python
df.to_csv("cleaned_spotify_data.csv", index=False)
print("Saved successfully!")
```
